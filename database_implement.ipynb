# Start Creating the RDBMS for the apps collected:

[1] Create the companies table:

```python
sql_1 = ''' SELECT DISTINCT	app_company FROM appsTableOne'''
col = connection.execute(sql_1).fetchall()
```

```python
DB2_Con = sqlite3.connect("apps_RDBMS.db")
cur2 = DB2_Con.cursor()
sql_2 = cur2.executemany('''INSERT INTO company (company_name) VALUES (?)''', col)
cur2.execute("SELECT * FROM company").fetchall()
DB2_Con.commit()
```

```python
companyk = []
for i in df1['app_company'].values:
  sql = "SELECT company_id FROM company WHERE company_name == ?"
  data = [i]
  companyk.append(cur2.execute(sql, data).fetchall())
  DB2_Con.commit()

for i in range(len(companyk)):
  for j in companyk[i][0]:
    companyk[i]=j
companyk
```

[2] create geners table

```python
sql_1 = ''' SELECT DISTINCT	app_gener FROM appsTableOne'''
col = connection.execute(sql_1).fetchall()
```

```python
sql_2 = cur2.executemany('''INSERT INTO geners (gener) VALUES (?)''', col)
res = cur2.execute("SELECT * FROM geners").fetchall()
DB2_Con.commit()
res
```

```python
generk = []
for i in df1['gener'].values:
  sql = "SELECT gener_id FROM geners WHERE gener == ?"
  data = [i]
  generk.append(cur2.execute(sql, data).fetchall())
  DB2_Con.commit()

for i in range(len(generk)):
  for j in generk[i][0]:
    generk[i]=j
generk
```

[3] create interactive_element table:

```python
sql_1 = ''' SELECT DISTINCT	interactive_elements FROM appsTableOne'''
col = connection.execute(sql_1).fetchall()
col
```

```python
sql_2 = cur2.executemany('''INSERT INTO interactive_elements (element) VALUES (?)''', col)
res = cur2.execute("SELECT * FROM interactive_elements").fetchall()
DB2_Con.commit()
res
```

```python
elementk = []
for i in df1['InterActive_Elem'].values:
  sql = "SELECT element_id FROM interactive_elements WHERE element == ?"
  data = [i]
  elementk.append(cur2.execute(sql, data).fetchall())
  DB2_Con.commit()

for i in range(len(elementk)):
  if len(elementk[i])>0:
    for j in elementk[i][0]:
      elementk[i]=j
elementk
```

```python
for i in range(len(elementk)):
  if elementk[i]==[]:
    elementk[i] = None
```

```python
sql_2 = cur2.executemany('''INSERT INTO content (content_category) VALUES (?)''', col)
res = cur2.execute("SELECT * FROM content").fetchall()
DB2_Con.commit()
res
```

```python
contentk = []
for i in df1['content_Rating'].values:
  sql = "SELECT content_cat_id FROM content WHERE content_category == ?"
  data = [i]
  contentk.append(cur2.execute(sql, data).fetchall())
  DB2_Con.commit()

for i in range(len(contentk)):
  if len(contentk[i])>0:
    for j in contentk[i][0]:
      contentk[i]=j
contentk
```

```python
sql_2 = cur2.executemany('''INSERT INTO reports (report_type) VALUES (?)''', col)
res = cur2.execute("SELECT * FROM reports").fetchall()
DB2_Con.commit()
res
```

```python
reportk = []
for i in df1['Report'].values:
  sql = "SELECT report_type_id FROM reports WHERE report_type == ?"
  data = [i]
  reportk.append(cur2.execute(sql, data).fetchall())
  DB2_Con.commit()

for i in range(len(reportk)):
  if len(reportk[i])>0:
    for j in reportk[i][0]:
      reportk[i]=j

reportk
```

[6] create pages table

```python
sql_1 = ''' SELECT DISTINCT	page_index, app_under_title  FROM appsTableOne'''
col = connection.execute(sql_1).fetchall()
col
```

```python
sql_2 = cur2.executemany('''INSERT INTO pages (page_appearance_order, page_name) VALUES (?,?)''', col)
res = cur2.execute("SELECT * FROM pages").fetchall()
DB2_Con.commit()
res
```

```python
titlek = []
for i in df1['underTitle'].values:
  sql = "SELECT page_id FROM pages WHERE page_name == ?"
  data = [i]
  titlek.append(cur2.execute(sql, data).fetchall())
  DB2_Con.commit()

for i in range(len(titlek)):
  if len(titlek[i])>0:
    for j in titlek[i][0]:
      titlek[i]=j

titlek
```

[7] create main app table:

```python
sql_1 = ''' SELECT app_name, app_rating, app_size, app_price, no_of_comments, no_of_installs,
inapp_purchases, latest_update, android_versions_supported FROM appsTableOne'''
col = connection.execute(sql_1).fetchall()
col
```

```python
d = []

for i,t,r,c,e,g,co in zip(range(len(col)),titlek,reportk,contentk,elementk,generk,companyk):
  in_d =[]
  for j in col[i]:
    in_d.append(j)
  in_d.append(t)
  in_d.append(r)
  in_d.append(c)
  in_d.append(e)
  in_d.append(g)
  in_d.append(co)

  d.append(tuple(in_d))
d
```

```python
sql_2 = cur2.executemany('''INSERT INTO apps 
(app_name, app_rating, app_size, app_price, no_of_comments, no_of_installs, 
in_app_purchases, latest_update, android_version_supported, page_id, 
report_type_id, content_cat_id, element_id, gener_id, comapny_id)
VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)''', d)
res = cur2.execute("SELECT * FROM apps").fetchall()
DB2_Con.commit()
res
```
